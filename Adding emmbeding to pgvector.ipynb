```python
from langchain.vectorstores import Chroma
from langchain.text_splitter import RecursiveCharacterTextSplitter,CharacterTextSplitter,TokenTextSplitter
from langchain.chains import RetrievalQA
from langchain.document_loaders import PyPDFLoader, Docx2txtLoader
from langchain.document_loaders import DirectoryLoader
from langchain.chains import RetrievalQA


from InstructorEmbedding import INSTRUCTOR
from langchain.embeddings import HuggingFaceInstructEmbeddings
```

```python
# load pdfs
loader = DirectoryLoader('./data/', glob="./*.pdf", loader_cls=PyPDFLoader)
documents = loader.load()
```

```python
# load docx
loader_docs = DirectoryLoader('./data/', glob="./*.docx", loader_cls=Docx2txtLoader)
documents_docs = loader_docs.load()
```

```python
print(documents[0].page_content)
```

```python
len(documents)
```

```python
# pdf
text_splitter = TokenTextSplitter( chunk_size=1000, chunk_overlap=100 )
texts = text_splitter.split_documents(documents)

# docx
texts_docs = text_splitter.split_documents(documents_docs)

all_text = texts + texts_docs
```

```python
all_text[-1].page_content 
```

```python
# Embedding
instructor_embeddings = HuggingFaceInstructEmbeddings(model_name="hkunlp/instructor-xl",
                                                      # model_kwargs={"device": "cuda"}
                                                     )
```

```python
# takes hours in cpu to get the embeding of all chunks so i take just one document instead

all_text = texts_docs
len(all_text)
```

```python
from langchain.vectorstores.pgvector import PGVector

CONNECTION_STRING = "postgresql+psycopg2://postgres:test@localhost:5432/vector_db"'
```

```python


COLLECTION_NAME = "state_of_the_union_test"

db = PGVector.from_documents(
    embedding=instructor_embeddings,
    documents=all_text,
    collection_name=COLLECTION_NAME,
    connection_string=CONNECTION_STRING,
)
```

```python

query = "Seçimlerin genel yönetim"
similar = db.similarity_search_with_score(query, k=5)

for doc in similar:
    print(doc[0].page_content, end="\n\n")
    print( "#"* 100 + "\n\n")
```

```python

# read db
store = PGVector(
    collection_name=COLLECTION_NAME,
    connection_string=CONNECTION_STRING,
    embedding_function= instructor_embeddings,
)
 
```

```python
query = "Seçimlerin genel yönetim"
similar = store.similarity_search_with_score(query, k=5)

for doc in similar:
    print(doc[0].page_content, end="\n\n")
    print( "#"* 100 + "\n\n")
```
